# Labeling Postprocessor

## Dependencies

```python
import os
import shutil
```

## Parameters

```python
input_root = "AI.R-Multimodal_Thermal_Dataset.v1i.yolov8"
input_sub = "train"
input_path = os.path.join(input_root, input_sub)
input_images_path = os.path.join(input_path, "images")
input_labels_path = os.path.join(input_path, "labels")
output_root = "After labelling"
main_output_folders = ["cloudy", "darkness", "glare", "misc", "snowy"]
```

## Copy images in right folder

```python
def organize_images(input_images_path, output_root):
    i = 0
    for filename in os.listdir(input_images_path):
        # if i > 3:
        #     continue  # Limit to first files for testing
        i += 1
        if not filename.endswith('.jpg'):
            continue
        parts = filename.split('_')
        main_folder = parts[0]
        print(main_folder)


        if main_folder not in main_output_folders:
                    print(f"Dossier principal inconnu pour {filename}, ignoré.")
                    continue

        subfolder_name = parts[1] + "_" + parts[2]
        print(subfolder_name)

        out_filename = ""
        for part in parts[3:-1]:
            out_filename += part + "_"
        out_filename = out_filename[:-1] # Remove the last underscore
        out_filename += ".jpg"
        print(out_filename)
        
        output_subfolder = os.path.join(output_root, main_folder, subfolder_name)
        os.makedirs(output_subfolder, exist_ok=True)
        file_path = os.path.join(input_images_path, filename)

        shutil.copy2(file_path, os.path.join(output_subfolder, out_filename))
```

## Copy labels in right folder

```python
def organize_labels(input_labels_path, output_root):
    i = 0
    for filename in os.listdir(input_labels_path):
        # if i > 3:
        #     continue  # Limit to first files for testing
        i += 1
        if not filename.endswith('.txt'):
            continue
        parts = filename.split('_')
        main_folder = parts[0]
        print(main_folder)


        if main_folder not in main_output_folders:
                    print(f"Dossier principal inconnu pour {filename}, ignoré.")
                    continue

        subfolder_name = parts[1] + "_" + parts[2]
        print(subfolder_name)

        out_filename = ""
        for part in parts[3:-1]:
            out_filename += part + "_"
        out_filename = out_filename[:-1] # Remove the last underscore
        out_filename += ".txt"
        print(out_filename)
        
        output_subfolder = os.path.join(output_root, main_folder, subfolder_name)
        os.makedirs(output_subfolder, exist_ok=True)
        file_path = os.path.join(input_labels_path, filename)

        shutil.copy2(file_path, os.path.join(output_subfolder, out_filename))
```

## Loop to all files available in folders

```python
for folder_name in os.listdir(input_path):
    folder_path = os.path.join(input_path, folder_name)
    if(folder_path == input_images_path):
        organize_images(input_images_path, output_root)
    if(folder_path == input_labels_path):
        organize_labels(input_labels_path, output_root)
```
