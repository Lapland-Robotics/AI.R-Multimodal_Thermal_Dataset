# File Ordering

## Dependencies

```python
import os
import shutil
```

## Parameters

```python
input_root = "labels"
input_sub = "train"
output_root = "exemplary_subset_new"
```

## Rename images correctly

```python
def rename_image(filename, folder_path):
    parts = filename.split('_')

    out_filename = ""
    for part in parts[3:-1]:
        out_filename += part + "_"
    out_filename = out_filename[:-1] # Remove the last underscore
    out_filename += ".jpg"
    # print(out_filename)

    os.rename(os.path.join(folder_path, filename), os.path.join(folder_path, out_filename))

    return
```

## Copy labels in right folder

```python
def order_file(filename):
    if not filename.endswith('.txt'):
            return
    parts = filename.split('_')

    subfolder_name = parts[1] + "_" + parts[2]
    # print(subfolder_name)

    out_filename = ""
    for part in parts[3:-1]:
        out_filename += part + "_"
    out_filename = out_filename[:-1] # Remove the last underscore
    out_filename += ".txt"
    # print(out_filename)

    output_folder = os.path.join(output_root, subfolder_name)
    os.makedirs(output_folder, exist_ok=True)

    file_path = os.path.join(input_root, filename)
    shutil.copy2(file_path, os.path.join(output_folder,out_filename))
```

## Loop to all files available in folders

```python
i = 0
for folder_name in os.listdir(output_root):
    # if i > 1:
    #     continue  # Limit to first folders for testing

    i = i + 1
    print("Processing folder : " + folder_name)

    folder_path = os.path.join(output_root, folder_name)
    for filename in os.listdir(folder_path):
        if not filename.endswith('.jpg'):
            continue
        rename_image(filename, folder_path)

i = 0
for filename in os.listdir(input_root):
    # if i > 1:
    #     continue  # Limit to first files for testing

    i = i + 1
    print("Processing file : " + filename)

    if not filename.endswith('.txt'):
            continue
    
    file_path = os.path.join(input_root, filename)
    order_file(filename)

    

```
