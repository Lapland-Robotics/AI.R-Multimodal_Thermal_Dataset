# Labeling Preprocessor

## Dependencies

```python
import os
import shutil
```

## Parameters

```python
input_root = "."
output_root = "After name processing before labeling"
```

## Copy images in right folder

```python
def CopyFiles(folder_path, output_root):
    # Prepare output folder
    os.makedirs(output_root, exist_ok=True)

    # Define paths for the images
    left_rgb_path = os.path.join(folder_path, "left_rgb.png")
    scaled_left_rgb_path = os.path.join(folder_path, "scaled_left_rgb.png")
    thermal_path = os.path.join(folder_path, "thermal.png")
    
    # Copy images as is
    shutil.copy2(left_rgb_path, os.path.join(output_root, "left_rgb.png"))
    shutil.copy2(scaled_left_rgb_path, os.path.join(output_root, "scaled_left_rgb.png"))
    shutil.copy2(thermal_path, os.path.join(output_root, "thermal.png"))

    print(f"    Copied images from {folder_path} to {output_root}")
```

## Rename Image


```python
def RenameFiles(folder_name, folder_subname, output_root):
    # Rename the images in the output folder
    base_name = folder_name + "_" + folder_subname + "_"
    os.rename(os.path.join(output_root, "left_rgb.png"), os.path.join(output_root, base_name + "left_rgb.png"))
    os.rename(os.path.join(output_root, "scaled_left_rgb.png"), os.path.join(output_root, base_name + "scaled_left_rgb.png"))
    os.rename(os.path.join(output_root, "thermal.png"), os.path.join(output_root, base_name +  "thermal.png"))    
```

## Loop to all files available in folders

```python
i = 0
for folder_name in os.listdir(input_root):
    if (folder_name == output_root):
        continue  # Skip the output folder itself
    print("Processing folder: " + folder_name)
    folder_path = os.path.join(input_root, folder_name)
    if not os.path.isdir(folder_path):
        continue  # Skip files
    for folder_subname in os.listdir(folder_path):
        # if (i > 0):
        #     continue # Limit to first folders for testing
        print("Processing folder: " + folder_subname)
        folder_subpath = os.path.join(folder_path, folder_subname)

        # left_rgb_path = os.path.join(folder_path, "left_rgb.png")
        # scaled_left_rgb_path = os.path.join(folder_path, "scaled_left_rgb.png")
        # thermal_path = os.path.join(folder_path, "thermal.png")

        CopyFiles(folder_subpath, output_root)
        RenameFiles(folder_name, folder_subname, output_root)
        i+= 1
        print("Processed folder:"+str(i)+"/60")
print("Processing complete!")
```
